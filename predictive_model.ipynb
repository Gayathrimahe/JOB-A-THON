```python
def dataset_analysis(train_data, test_data):
    print("shape of test dataset",test_data.shape)
    print("shape of train dataset", train_data.shape)
    print("columns of test dataset", test_data.columns)
    print("columns of train dataset", train_data.columns)
    print(train_data.describe())
    print(test_data.describe())
    print(train_data.isnull().values.any())
    print(test_data.isnull().values.any())
    
```

```python
dataset_analysis(train_data, test_data)
```
